```python
import numpy as np
#lav data ordinal

# Saml alle data (globalt)
all_data = {**europe_data, **africa_data, **middle_east_data, **asia_data, **south_america_data}

# Beregn median for hvert land
median_salaries = [(min_s + max_s) / 2 for (min_s, max_s) in all_data.values()]

# Beregn 25. og 75. percentil
p25 = np.percentile(median_salaries, 25)
p75 = np.percentile(median_salaries, 75)

# Kategoriser hvert land
categorized = {}
for country, (min_s, max_s) in all_data.items():
    median = (min_s + max_s) / 2
    if median < p25:
        cat = 1  # Lav
    elif median <= p75:
        cat = 2  # Mellem
    else:
        cat = 3  # Høj
    categorized[country] = cat

# Print grænser
print("Global kategorisering baseret på median:")
print(f"- Lav (1): Median < {p25:.0f}")
print(f"- Mellem (2): {p25:.0f} ≤ Median ≤ {p75:.0f}")
print(f"- Høj (3): Median > {p75:.0f}\n")

# Eksempel på output
print("Eksempel på kategorisering (første 5 lande):")
for country in list(categorized.keys())[:5]:
    print(f"{country}: {all_data[country]} → Median = {(all_data[country][0] + all_data[country][1]) / 2:.0f} → Kategori {categorized[country]}")
```

```python
import numpy as np
import collections

# Jeres region-data
regions = {
    "Europe": europe_data,
    "Africa": africa_data,
    "Middle East": middle_east_data,
    "Asia": asia_data,
    "South America": south_america_data
}

print("Summary statistics per region (based on ordinal salary categories):\n")

for region_name, region_dict in regions.items():
    # Tag kun kategorien for landene i den aktuelle region
    region_categories = [categorized[country] for country in region_dict]

    # Frekvenser og procentfordeling
    counts = collections.Counter(region_categories)
    total = len(region_categories)
    percentages = {cat: (counts[cat] / total) * 100 for cat in [1, 2, 3]}

    # Median og spænd
    median_cat = int(np.median(region_categories))
    min_cat = min(region_categories)
    max_cat = max(region_categories)

    print(f"Region: {region_name}")
    for cat in [1, 2, 3]:
        label = {1: "Lav", 2: "Mellem", 3: "Høj"}[cat]
        print(f"  - {label} ({cat}): {counts[cat]} lande ({percentages[cat]:.1f}%)")
    print(f"  Median kategori: {median_cat} ({ {1: 'Lav', 2: 'Mellem', 3: 'Høj'}[median_cat] })")
    print(f"  Spænde: {min_cat} til {max_cat}\n")
```

```python
import numpy as np
from scipy import stats

#kruskal på ordinal data:


# Opret en dictionary med de kategoriserede data (fra tidligere)
categorized_data = {
    "Europe": [categorized[country] for country in europe_data.keys()],
    "Africa": [categorized[country] for country in africa_data.keys()],
    "Middle East": [categorized[country] for country in middle_east_data.keys()],
    "Asia": [categorized[country] for country in asia_data.keys()],
    "South America": [categorized[country] for country in south_america_data.keys()],
}

# Udfør Kruskal-Wallis på de ordinale kategorier
statistic, p_value = stats.kruskal(*categorized_data.values())


print("Kruskal-Wallis Test på Ordinale Kategorier (1=lav, 2=mellem, 3=høj):")
print(f"Test-statistik: {statistic:.3f}")
print(f"P-værdi: {p_value:.5f}")

if p_value < 0.05:
    print("\nKonklusion: Signifikant forskel i lønintervalfordeling mellem regioner (p < 0.05).")
    print("Brug post-hoc test (f.eks. Dunn's test) for at finde hvilke regioner der adskiller sig.")
else:
    print("\nKonklusion: Ingen signifikant forskel i lønintervalfordeling (p > 0.05).")
```

```python
import scikit_posthocs as sp
import pandas as pd

# Konverter data til long-format
data = []
groups = []

for region, values in categorized_data.items():
    data.extend(values)
    groups.extend([region] * len(values))

df = pd.DataFrame({'Region': groups, 'Score': data})

# Dunn's test med Holm-korrektion
dunn_results_holm = sp.posthoc_dunn(df, val_col='Score', group_col='Region', p_adjust='holm')

print("\nDunn's Test med Holm-korrektion:")
print(dunn_results_holm)

import seaborn as sns
import matplotlib.pyplot as plt

# Lav et heatmap af p-værdierne med 4 decimaler
plt.figure(figsize=(10, 8))
sns.heatmap(dunn_results_holm, annot=True, fmt=".4f", cmap="coolwarm", cbar_kws={'label': 'P-værdi'})
plt.title("Heatmap af Dunn's Test (Holm-korrektion)")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

```
