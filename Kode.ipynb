### Summary statiscs

```python
import numpy as np
from scipy.stats import skew, shapiro, pearsonr, spearmanr
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pandas as pd

# Function to calculate median values from ranges
def get_median_values(data_dict):
    return np.array([np.median([low, high]) for _, (low, high) in data_dict.items()])

# Calculate median values for each region
regions = {
    'Europe': europe_data,
    'Africa': africa_data,
    'Middle East': middle_east_data,
    'Asia': asia_data,
    'South America': south_america_data
}

region_medians = {name: get_median_values(data) for name, data in regions.items()}

def print_descriptive_stats(data, name):
    print(f"\n--- Descriptive statistics for {name} ---")
    print(f"Count: {len(data):,}")
    print(f"Mean: {np.mean(data):,.2f}")
    print(f"Median: {np.median(data):,.2f}")
    print(f"Std dev: {np.std(data, ddof=1):,.2f}")
    print(f"Min: {np.min(data):,.2f}")
    print(f"Max: {np.max(data):,.2f}")
    print(f"25th percentile: {np.percentile(data, 25):,.2f}")
    print(f"75th percentile: {np.percentile(data, 75):,.2f}")

# Print stats for all regions
for name, data in region_medians.items():
    print_descriptive_stats(data, name)

# Skewness and variance
def print_skew_var(data, name):
    print(f"\n--- Shape analysis for {name} ---")
    print(f"Skewness: {skew(data):.4f}")
    print(f"Variance: {np.var(data, ddof=1):,.2f}")

for name, data in region_medians.items():
    print_skew_var(data, name)

# Comparative plots
plt.figure(figsize=(20, 12))

# Histograms
for i, (name, data) in enumerate(region_medians.items(), 1):
    plt.subplot(2, 3, i)
    sns.histplot(data, bins=15, kde=True)
    plt.title(f'{name} Median Values')
    plt.xlabel('Values')

plt.tight_layout()
plt.show()

# Boxplots
plt.figure(figsize=(20, 6))
sns.boxplot(data=pd.DataFrame({k: pd.Series(v) for k, v in region_medians.items()}))
plt.title('Comparative Median Values Across Regions')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Q-Q plots
plt.figure(figsize=(20, 12))

for i, (name, data) in enumerate(region_medians.items(), 1):
    plt.subplot(2, 3, i)
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(f'Q-Q plot: {name}')

plt.tight_layout()
plt.show()

# Shapiro-Wilk tests
def print_shapiro(data, name):
    w, p = shapiro(data)
    print(f"\n--- Normality test for {name} ---")
    print(f"Shapiro-Wilk p-value = {p:.4f}")
    print("Conclusion:", "✅ Normally distributed" if p > 0.05 else "❌ Not normally distributed")

for name, data in region_medians.items():
    print_shapiro(data, name)

# Correlation analysis
region_names = list(region_medians.keys())
n_regions = len(region_names)
n_pairs = n_regions * (n_regions - 1) // 2

plt.figure(figsize=(18, 5 * ((n_pairs + 2) // 3)))
pair_count = 0

for i in range(n_regions):
    for j in range(i + 1, n_regions):
        pair_count += 1
        cont1 = region_names[i]
        cont2 = region_names[j]
        
        # Get data and match lengths
        x_full = region_medians[cont1]
        y_full = region_medians[cont2]
        min_len = min(len(x_full), len(y_full))
        x = x_full[:min_len]
        y = y_full[:min_len]
        
        # Calculate correlations
        pearson_r, pearson_p = pearsonr(x, y)
        spearman_rho, spearman_p = spearmanr(x, y)
        
        # Create scatter plot
        plt.subplot((n_pairs + 2) // 3, 3, pair_count)
        sns.regplot(x=x, y=y, ci=None, scatter_kws={'alpha':0.6, 's':80})
        plt.title(f'{cont1} vs {cont2} (n={min_len})')
        plt.xlabel(f'{cont1} Values')
        plt.ylabel(f'{cont2} Values')
        
        # Add correlation annotations
        plt.annotate(
            f"Pearson r = {pearson_r:.2f}\n(p={pearson_p:.3f})\n\n"
            f"Spearman ρ = {spearman_rho:.2f}\n(p={spearman_p:.3f})",
            xy=(0.05, 0.75), xycoords='axes fraction',
            bbox=dict(boxstyle='round', fc='white', alpha=0.8),
            fontsize=9
        )
        
        # Add y=x reference line
        lims = [min(x.min(), y.min()), max(x.max(), y.max())]
        plt.plot(lims, lims, 'r--', alpha=0.3, label='y=x')
        plt.legend()

plt.tight_layout()
plt.show()
```

### Kruskal-Wallis Test

```python
import scipy.stats as stats

# Funktion til at omregne interval til enkelt observation ved gennemsnit
def avg_interval(data):
    return [(low + high) / 2 for (low, high) in data.values()]

# Ekstraher observationer for hver region (gennemsnit af intervaller)
europe_obs = avg_interval(europe_data)
africa_obs = avg_interval(africa_data)
middle_east_obs = avg_interval(middle_east_data)
asia_obs = avg_interval(asia_data)
south_america_obs = avg_interval(south_america_data)

# Kør Kruskal-Wallis test
stat, p_value = stats.kruskal(europe_obs, africa_obs, middle_east_obs, south_america_obs, asia_obs)

print(f"Kruskal-Wallis test statistik: {stat}")
print(f"P-værdi: {p_value}")
```
